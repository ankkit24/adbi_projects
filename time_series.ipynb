```python
import numpy as np
import pandas as pd
from scipy import stats
import matplotlib.pyplot as plt
```

```python
# use 'conda install statsmodel' if the module is not found
import statsmodels.api as sm
```

```python
#import for seasonal plots
import statsmodels.tsa.seasonal as seasonal
from statsmodels.graphics.api import qqplot
```

```python
data=pd.read_csv("./Amtrak_data.csv")
data.columns
data=data[~np.isnan(data["Ridership"])]
data.index=pd.Index(sm.tsa.datetools.dates_from_range("1991m1","2004m3"))
del data["Month"]
```

```python
#we have taken out data for a time period
# now we plot it
plt.plot(data)
plt.show()
```

```python
#plotting the lag plots
for i in range(1,7):
    plt.subplot(2,3,i)
    pd.plotting.lag_plot(data,lag=i)
```

```python
#getting the autocorrelation using acf
sm.graphics.tsa.plot_acf(data,lags=12)
```

```python
plt.plot(np.diff(data["Ridership"],n=3))
plt.show()
```
